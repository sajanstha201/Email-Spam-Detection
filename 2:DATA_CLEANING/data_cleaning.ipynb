```python
data=pd.read_csv('/content/drive/MyDrive/Data Science/EMAIL SPAM DETECTION/DATA/email_subject_based.csv')
data=data.loc[:,['text','spam']]
data['text']=data['text'].str.strip()
data['text']=[string[13:] if string[:13]=='Subject: re :' else string[8:] for string in data['text']]
data=data.rename(columns={'text':'subject'})
data.dropna(inplace=True)
data
```

```python
data.to_csv('/content/drive/MyDrive/Data Science/EMAIL SPAM DETECTION/DATA/cleaned_email_subject_based.csv',index=False)
```

```python
data2=pd.read_csv('/content/drive/MyDrive/Data Science/EMAIL SPAM DETECTION/DATA/email_message_based.csv')
data2['Category']=[0 if e=='ham' else 1 for e in data2['Category']]
data2=data2.rename(columns={'Category':'spam'})
data2.dropna(inplace=True)
data2
```
