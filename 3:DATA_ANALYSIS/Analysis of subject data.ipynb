```python
df=pd.read_csv('/content/drive/MyDrive/Data Science/EMAIL SPAM DETECTION/DATA/cleaned_email_subject_based.csv')
data=pd.DataFrame(df)
```

```python
subject_0=0
subject_1=0
for i in data['spam']:
  if(i=='0'):
    subject_0=subject_0+1
  else:
    subject_1=subject_1+1
plt.bar([0],subject_0,width=1,label='ham')
plt.bar([1.5],subject_1,width=1,label='spam')
plt.xticks([0,1.5],['ham','spam'])
plt.legend()
plt.title('Number of Spam and Ham in Subject Data')
plt.ylabel('No of Data')
```

```python
data['length']=[len(e) for e in data['subject']]
unique_length_arr=np.sort(data['length'].unique())
spam_unique_arr=[]
ham_unique_arr=[]
for i in range(len(unique_length_arr)):
  spam_unique_arr.append(0)
  ham_unique_arr.append(0)
for i in range(data.shape[0]):
  if(data.iloc[i].spam=='0'):
    l=np.where(unique_length_arr == data.iloc[i].length)[0][0]
    ham_unique_arr[l] += 1
  else:
    spam_unique_arr[np.where(unique_length_arr==data.iloc[i].length)[0][0]]+=1
plt.bar(unique_length_arr[50:200],ham_unique_arr[50:200],label='ham',color='blue',alpha=0.5)
plt.bar(unique_length_arr[50:200],spam_unique_arr[50:200],label='spam',color='red',alpha=0.5)
plt.legend()
plt.xlabel('length of subject')
plt.ylabel('No of mail')
plt.title('Length of subject vs Frequency')
```
