```python
import pandas as pd
import numpy as np
import tensorflow as tf
from transformers import BertTokenizer,TFBertModel
```

```python
from google.colab import drive
drive.mount('/content/drive/')
```

```python
data=pd.read_csv('/content/drive/MyDrive/Data Science/EMAIL SPAM DETECTION/DATA/cleaned_email_subject_based.csv')
```

```python
data
```

```python
tokenizer=BertTokenizer.from_pretrained('bert-base-cased')
model=TFBertModel.from_pretrained('bert-base-cased')
```

```python
vectorized_subject=[]
for e in data['subject']:
  vector=model(tokenizer(e,return_tensors='tf',padding=True,truncation=True,max_length=256))
  vectorized_subject.append(vector)
```

```python
np.save('/content/drive/MyDrive/Data Science/EMAIL SPAM DETECTION/DATA/vectorized_subject_without_target.npy',vectorized_subject)
```

```python
l=[]
for i in range(len(data['spam'])):
  x=list(vectorized_subject[i])
  x.append(data['spam'][i])
  l.append(x)
```

```python
np.save('/content/drive/MyDrive/Data Science/EMAIL SPAM DETECTION/DATA/vectorized_subject_with_target.npy',l)
```
